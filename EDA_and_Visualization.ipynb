```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
breast_cancer_data = pd.read_csv('/content/data.csv')
```

```python
breast_cancer_data.head()
```

```python
breast_cancer_data.tail()
```

```python
breast_cancer_data.drop( columns= 'Unnamed: 32', axis = 1 , inplace = True)
```

```python
breast_cancer_data.shape
```

```python
breast_cancer_data.info()
```

```python
breast_cancer_data.drop( columns = 'id', axis = 1, inplace = True)
```

```python
breast_cancer_data.describe()   #distributive statisctics
```

```python
#checking the target vslues

breast_cancer_data['diagnosis'].value_counts()
```

```python
#encoding the tatrget columns

label_encoding = LabelEncoder()

labels = label_encoding.fit_transform(breast_cancer_data['diagnosis'])

breast_cancer_data['target'] = labels

breast_cancer_data.drop( columns = 'diagnosis', axis = 1, inplace = True)
```

```python
breast_cancer_data.head()
```

```python
breast_cancer_data['target'].value_counts()
```

```python
breast_cancer_data.groupby('target').mean()
```

```python
#countplot for the target columns for checking the distribution pf target
sns.countplot(x = 'target', data = breast_cancer_data)
```

```python
#distibution plot for all columns

for columns in breast_cancer_data:
  print(columns)
```

```python
#creating a for loop to get the distribution plot for all columns

for columns in breast_cancer_data:
  sns.displot(x=columns, data = breast_cancer_data)
```

```python
sns.distplot(x =breast_cancer_data.radius_mean)
```

```python
#scatter plot

first_columns = breast_cancer_data.iloc[:,0]

second_columns = breast_cancer_data.iloc[:, 1]
```

```python
plt.scatter(x= first_columns, y = second_columns)
```

```python
#Boxplot

for columns in breast_cancer_data:
  plt.figure()
  breast_cancer_data.boxplot(columns)


  #the black spots are outliers
```
